```python
#Importing libraries
import pandas as pd
```

```python
# 1 (2 marks). Load the netflix_titles.csv dataset as a Pandas dataframe: 
df = pd.read_csv('netflix_titles.csv')
df
```

```python
# 2 (2 marks). Show the first 10 rows of the dataset
df.head(10)
```

```python
# 3 (1 mark). Show the last 5 rows 
df.tail(5)
```

```python
# 4 (4 marks). Count the number of rows which have no missing value in any of the columns, and show the resulting data frame.

non_missing_rows = df.dropna().shape[0]

print("Number of rows without missing values:", non_missing_rows)

# Show the resulting DataFrame with no missing values
df_no_missing = df.dropna()
df_no_missing
```

```python
# 5 (3 marks). Drop everything that is not a 'Movie' and show the resulting data frame.
df = df[df['type'] == 'Movie']
print(df)
```

```python
df.head(10)
```

```python
# 6 (7 marks). For every movie where the duration is missing, replace it with the average 
  # duration of the movies in the dataset.


#extract only numeric values
df['duration'] = df['duration'].str.extract('(\d+)').astype(float)
average_duration = df['duration'].mean()
# Replace missing durations with the calculated average
df['duration'] = df['duration'].fillna(average_duration)
```

```python
df.head(50)
```

```python
df['duration'].mean()
```

```python
# 7 (3 marks). Are there any duplicates in the data? If yes, please remove them.
df = df.drop_duplicates()
```

```python
# 8 (2 marks). Rename all columns to start with a capital letter. 

# Create a dictionary to map current column names to capitalized versions
column_mapped = {col: col.capitalize() for col in df.columns}

# Rename columns with the capitalized names
df = df.rename(columns=column_mapped)
```

```python
df.head(50)
```

```python
# 9 (1 mark). Save the cleaned dataset as a new csv file. 
df.to_csv('df_cleaned.csv', index=False)
```

```python
# 1 (10 marks). Create a visualisation which shows the frequency of movies by year of release.

#I'll interprete that as how many movies for each year of release

#library already imported

# Group by release year and count the frequency
movies_by_year = df['Release_year'].value_counts().sort_index()
#or movies_by_year = df['Release_year'].value_counts()

# Create the bar plot
plt.figure(figsize=(15, 10)) #I increased the size for the index to be readable
movies_by_year.plot(kind='bar', color='green')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.title('Frequency of Movies by Year of Release')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 2 (15 marks). Create a pie chart which shows the 5 top most popular genres of the movies 
  # in the dataset.




# Split the 'listed_in' column to create a list of genres for each movie
genres_split = df['Listed_in'].str.split(', ')

# Count the frequency of each genre
genre_counts = genres_split.explode().value_counts()

# Select the top 5 most popular genres
top_genres = genre_counts.head(5)

# Create the pie chart
plt.figure(figsize=(5, 5))
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Most Popular Genres of Movies')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

# Show the plot
plt.show()

```

```python
# 2 (5 marks). How many odd numbers do we have in the array?
#I am finding numbers that are not divisible by 2
sum_of_odd_numbers = (arr % 2 != 0).sum()
sum_of_odd_numbers
#The answer is 3 
```

```python
# 3 (5 marks). Replace all odd numbers in arr with 999
arr[arr % 2 != 0] = 999
arr
```

```python
# 4 (8 marks). The above is a 1D array. Convert it into a 4D array and reverse its columns.

# Desired output
# > array([[560,  82,  56,   0],
# >        [ 10,  92, 999, 140],
# >        [ 26,  18, 999, 999],
# >        [196, 364, 240, 150]])

# Reshape the 1D array into a 4D array and reverse its columns
arr_4d = arr.reshape((4, 4, 1, 1))
arr_4d_reversed = np.flip(arr_4d, axis=1)

# Convert the 4D array to a 2D array for the desired output
desired_output = arr_4d_reversed.reshape((4, 4))

print(desired_output)
```
